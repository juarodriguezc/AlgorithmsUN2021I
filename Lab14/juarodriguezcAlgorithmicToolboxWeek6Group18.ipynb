## **I Maximum Amount of Gold Challenge**

C++

```python
%%writefile exercise1.cpp
#include <iostream>
#include <vector>

using std::vector;
using std::endl;
using std::cout;
using std::cin;

void initializeMatrix(vector<vector<int>> &matrix, int W, vector<int> w){
    matrix.resize(W+1);
    for(int i = 0; i <= W; i++){
        matrix[i].resize(w.size()+1);
    }
    for(int i = 0; i <= W; i++){
        matrix[i][0] = 0;
    }
    for(int j = 0; j <= w.size(); j++){
        matrix[0][j] = 0;
    }
}
void printMatrix(vector<vector<int>> &matrix){
    cout<<"Column size: "<<matrix[0].size()<<endl;
    cout<<"Row size: "<<matrix.size()<<endl;
    for(int j=0; j<matrix[0].size();j++){
        for(int i=0; i<matrix.size();i++){
            cout<<matrix[i][j]<<" ";
        }
        cout<<endl;
    }
}
int optimal_weight(int W, vector<int> &w) {
    vector<vector <int>> value;
    initializeMatrix(value,W,w);
    for(int i = 1; i <= w.size(); i++){
        for(int j = 1; j <= W; j++){
            value[j][i] = value[j][i-1];
            if (w[i-1] <= j){
                int val = value [j - w[i-1]][i-1]+ w[i-1];
                if(value[j][i] < val){
                    value[j][i] = val;
                }
            }
        }
    }
    return value[W][w.size()];
}

int main() {
    int n, W;
    std::cin >> W >> n;
    vector<int> w(n);
    for (int i = 0; i < n; i++) {
        std::cin >> w[i];
    }
    std::cout << optimal_weight(W, w) << '\n';
}
```

## **III Solving the Maximum Value of an Arithmetic Expression Challenge**

C++

```python
%%writefile exercise3.cpp
#include <iostream>
#include <cassert>
#include <string>
#include <vector>
#include <stdlib.h>
#include <climits>

using std::string;
using std::max;
using std::min;
using std::vector;
using namespace std;

struct MinMaxStruct{
    long long m;
    long long M;
};
void initializeMatrix(vector <vector <long long> > &matrix, int nSize){
    matrix.resize(nSize+1);
    for(int a=0; a<matrix.size();a++){
        matrix[a].resize(nSize+1);
    }
    for(int a=0; a<matrix.size();a++){
        matrix[a][0] = 0;
    }
    for(int a=0; a<matrix[0].size();a++){
        matrix[0][a] = 0;
    }
}
long long eval(long long a, long long b, char op) {
    if (op == '*') {
        return a * b;
    }else if (op == '+') {
        return a + b;
    }else if (op == '-') {
        return a - b;
    }else {
        assert(0);
    }
}
MinMaxStruct MinAndMax(int i, int j, vector< vector <long long> > &m, vector< vector <long long> > &M, vector <char> &op){
    MinMaxStruct result;
    long long minN = LLONG_MAX;
    long long maxN = LLONG_MIN;
    for( int k = i; k <= j-1; k++){
        long long a = eval (M[k][i], M[j][k+1], op[k-1]);
        long long b = eval (M[k][i], m[j][k+1], op[k-1]);
        long long c = eval (m[k][i], M[j][k+1], op[k-1]);
        long long d = eval (m[k][i], m[j][k+1], op[k-1]);
        minN = min(minN, min(a, min(b, min(c, d))));
        maxN = max(maxN, max(a, max(b, max(c, d))));
    }
    result.m = minN;
    result.M = maxN;
    return result;
}
long long get_maximum_value(const string &exp) {
    int expSize = exp.size();
    int nExp = (expSize+1)/2;
    vector < vector <long long> > m;
    vector < vector <long long> > M;
    vector <char> op (expSize-nExp);
    initializeMatrix(m, nExp);
    initializeMatrix(M, nExp);
    {
        int j = 1;
        int k = 0;
        for(int i = 0; i < expSize; i++){
            if(exp[i] != '*' && exp[i] != '+' && exp[i] != '-'){
                m[j][j] = (long long)exp[i] - '0';
                M[j][j] = (long long)exp[i] - '0';
                j++;
            }else{
                op[k] = exp[i];
                k++;
            }
        }
    }
    for (int s = 1; s <= (nExp-1); s++ ){
        for(int i = 1; i <= (nExp -s); i++){
            int j = i + s;
            MinMaxStruct res = MinAndMax(i, j, m, M, op);
            m[j][i] = res.m;
            M[j][i] = res.M;
        }
    }
    return M[nExp][1];
}

int main() {
    string s;
    std::cin >> s;
    std::cout << get_maximum_value(s) << '\n';
}


```
