## **I Solving the Money Change Challenge**

C++

```python
%%writefile exercise1.cpp
#include <iostream>
using namespace std;
int get_change(int m) {
    int monedas[3]={10,5,1};
    int c_monedas = 0;
    //Repetir 3 veces por la cantidad de monedas
    for(int a=0; a<3; a++){
        if(m == 0) return c_monedas;
        int monedasporvalor = (m-(m%monedas[a]))/monedas[a];
        c_monedas += monedasporvalor;
        m = m - (monedasporvalor*monedas[a]);
    }
    return c_monedas;
}

int main() {
  int m;
  std::cin >> m;
  std::cout << get_change(m) << '\n';
}
```

## **II Solving the Maximum Value of the Loot Challenge**

C++

```python
%%writefile exercise2.cpp
#include <iostream>
#include <vector>
using std::vector;
using namespace std;
double get_optimal_value(int capacity, vector<int> weights, vector<int> values) {
    double value = 0.0;
    int space_left = capacity;
    for(int a=0; a<values.size();a++){
        if(space_left == 0)return value;
        int index_max = 0;
        for(int b=0; b<weights.size();b++){
            if(weights[b]>0){
                if(values[b]/(weights[b]+0.0)>values[index_max]/(weights[index_max]+0.0))
                    index_max = b;
            }
        }
        int weight_value = (space_left < weights[index_max])? space_left : weights[index_max];
        value += weight_value * (values[index_max]/(weights[index_max]+0.0));
        space_left -= weight_value;
        weights[index_max] -= weight_value;
    }
    return value;
}

int main() {
  int n;
  int capacity;
  std::cin >> n >> capacity;
  vector<int> values(n);
  vector<int> weights(n);
  for (int i = 0; i < n; i++) {
    std::cin >> values[i] >> weights[i];
  }

  double optimal_value = get_optimal_value(capacity, weights, values);

  std::cout.precision(10);
  std::cout << optimal_value << std::endl;
  return 0;
}
```

## **III Solving the Car Fueling Challenge**

C++

```python
%%writefile exercise3.cpp
#include <iostream>
#include <vector>

using std::cin;
using std::cout;
using std::vector;
using std::max;
using std::endl;


int compute_min_refills(int dist, int tank, vector<int> & stops) {
    int posicion_actual = 0;
    int num_tanqueos = 0;
    while(posicion_actual < stops.size()-1){
        int ultimo_tanqueo = posicion_actual;
        while(ultimo_tanqueo < stops.size()-1 && (stops[ultimo_tanqueo+1]-stops[posicion_actual])<=tank){
            ultimo_tanqueo++;
        }
        if(stops[ultimo_tanqueo] == stops[posicion_actual])return -1;
        posicion_actual = ultimo_tanqueo;
        if(posicion_actual == stops.size()-1)return num_tanqueos;
        num_tanqueos++;
    }
    return num_tanqueos;
}


int main() {
    int d = 0;
    cin >> d;
    int m = 0;
    cin >> m;
    int n = 0;
    cin >> n;

    vector<int> stops(n+2);
    stops[0] = 0;
    for (size_t i = 1; i < n+1; ++i)
        cin >> stops.at(i);
    stops[n+1] = d;
    cout << compute_min_refills(d, m, stops) << "\n";

    return 0;
}
```

## **IV Solving the Maximum Advertisement Revenue Challenge**

C++

```python
%%writefile exercise4.cpp
#include <algorithm>
#include <iostream>
#include <vector>
using std::vector;
using std::cout;
using std::endl;


long long max_dot_product(vector<int> a, vector<int> b) {
    int n = a.size();
    long long prod = 0;
    for(int i=0; i<n; i++){
        int temp_may_a=0, temp_may_b=0;
        int temp_min_a=0, temp_min_b=0;
        for(int j = 0; j < n; j++){
            if(a[j]!=100001){
                temp_may_a = j;
                temp_min_a = j;
            }
            if(b[j]!=100001){
                temp_may_b = j;
                temp_min_b = j;
            }
            if(a[temp_may_a] != 100001 && b[temp_may_b]!= 100001)break;
        }
        for(int j=0; j<n;j++){
            temp_may_a = (a[j]!=100001&&(a[j] > a[temp_may_a]))?j:temp_may_a;
            temp_may_b = (b[j]!=100001&&(b[j] > b[temp_may_b]))?j:temp_may_b;
            temp_min_a = (a[j]!=100001&&(a[j] < a[temp_min_a]))?j:temp_min_a;
            temp_min_b = (b[j]!=100001&&(b[j] < b[temp_min_b]))?j:temp_min_b;
        }
        if((a[temp_may_a] * b[temp_may_b]) > (a[temp_min_a] * b[temp_min_b])){
            prod += (long long)a[temp_may_a] * b[temp_may_b];
            a[temp_may_a] = 100001;
            b[temp_may_b] = 100001;
        }else{
            prod += (long long)a[temp_min_a] * b[temp_min_b];
            a[temp_min_a] = 100001;
            b[temp_min_b] = 100001;
        }
    }
    return prod;
}

int main() {
  size_t n;
  std::cin >> n;
  vector<int> a(n), b(n);
  for (size_t i = 0; i < n; i++) {
    std::cin >> a[i];
  }
  for (size_t i = 0; i < n; i++) {
    std::cin >> b[i];
  }
  std::cout << max_dot_product(a, b) << std::endl;
}

```

## **V Solving the  Collecting Signatures Challenge**

C++

```python
%%writefile exercise5.cpp
#include <algorithm>
#include <iostream>
#include <climits>
#include <vector>


#include <algorithm>
#include <iostream>

using std::vector;
using std::cout;
using std::endl;
struct Segment {
    int start, end;
};
vector<int> optimal_points(vector<Segment> &segments) {
    vector<int> points;
    vector <Segment> seg_copy = segments;
    //Ordenar los segmentos respecto al punto de finalizacion
    int n = seg_copy.size();
    for (int i = 0; i < n - 1; i++){
        for (int j = 0; j < n - i - 1; j++){
            if (seg_copy[j].end > seg_copy[j + 1].end){
                Segment temp = seg_copy[j];
                seg_copy[j] = seg_copy[j + 1];
                seg_copy[j + 1] = temp;
            }
        }
    }
    vector<bool> puntos(n);
    for(int a=0; a<n; a++){
        if(puntos[a] == false){
            puntos[a] = true;
            points.push_back(seg_copy[a].end);
            for(int b = a+1; b<n; b++){
                if(seg_copy[a].end <= seg_copy[b].end && seg_copy[a].end >= seg_copy[b].start){
                    puntos[b]=true;
                }
                else{
                    break;
                }
            }
        }
    }
    return points;
}


int main() {
    unsigned int n;
    std::cin >> n;
    vector<Segment> segments(n);
    for (size_t i = 0; i < segments.size(); ++i) {
        std::cin >> segments[i].start >> segments[i].end;
    }
    vector <int> points = optimal_points(segments);
    std::cout << points.size() << "\n";
    for (size_t i = 0; i < points.size(); ++i) {
        std::cout << points[i] << " ";
    }
}

```

## **VI Solving the Maximum Number of Prizes Challenge**

C++

```python
%%writefile exercise6.cpp
#include <iostream>
#include <vector>

using std::vector;

vector<int> optimal_summands(int n) {
    vector<int> summands;
    int n2 = n;
    int i = 1;
    while(n2 > 0){
        if(i+(i+1)<=n2){
            summands.push_back(i);
            n2 -= i;
            i++;
        }
        else{
            summands.push_back(n2);
            n2 = 0;
        }
    }
    return summands;
}

int main() {
  int n;
  std::cin >> n;
  vector<int> summands = optimal_summands(n);
  std::cout << summands.size() << '\n';
  for (size_t i = 0; i < summands.size(); ++i) {
    std::cout << summands[i] << ' ';
  }
}

```

## **VII Solving the Maximum Salary Challenge**

C++

```python
%%writefile exercise7.cpp
#include <algorithm>
#include <sstream>
#include <iostream>
#include <vector>
#include <string>

using std::vector;
using std::string;
using std::cout;
using std::endl;
using std::cin;
using namespace std;

bool esAmayorB(string a, string b){
    return (a+b > b+a);
}

string largest_number(vector<string> a) {
    string answer ="";
    int n = a.size();
    for(int b=0; b<n; b++){
        string max = "0";
        int indexMax = -1;
        for (int c=0; c<n; c++){
            if(esAmayorB(a[c],max)){
                max = a[c];
                indexMax = c;
            }
        }
        answer+=max;
        a[indexMax] = "0";
    }
    return answer;
}

int main() {
    int n;
    std::cin >> n;
    vector<string> a(n);
    for (size_t i = 0; i < a.size(); i++) {
    std::cin >> a[i];
    }
    std::cout << largest_number(a);
}


```
